There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend.

## Q3
 
Which single borough has the largest standard deviation in the combined SAT score?

- Save your results as a pandas DataFrame called largest_std_dev.
- The DataFrame should contain one row, with:
    - "borough" - the name of the NYC borough with the largest standard deviation of "total_SAT".
    - "num_schools" - the number of schools in the borough.
    - "average_SAT" - the mean of "total_SAT".
    - "std_SAT" - the standard deviation of "total_SAT".
- Round all numeric values to two decimal places.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()

# Start coding here...
# Add as many cells as you like...
```

## Q1 Answer

```python
# Filter schools data, subset, and sort according to "average_math" values in descending order 
best_math_schools = schools[schools["average_math"] >= 800 * 0.8][["school_name", "average_math"]].sort_values("average_math", ascending = False)
best_math_schools
```

```python
# Create new column "total_SAT" under schools dataframe
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]

# Sort and select top 10, then subset according to the requirements
top_10_schools = schools.sort_values("total_SAT", ascending = False).iloc[0:10][["school_name", "total_SAT"]]
top_10_schools
```

```python
# Group "total_SAT" data by "borough", calculate key stats, and round the values
schools_w_stats = schools.groupby("borough")["total_SAT"].agg(["count", "mean", "std"]).round(2)

# Rename key stats
schools_w_stats = schools_w_stats.rename(columns = {"count":"num_schools", "mean":"average_SAT","std":"std_SAT"})
schools_w_stats
```

```python
# Select the row with the highest standard deviation value
largest_std_dev = schools_w_stats[schools_w_stats["std_SAT"] == schools_w_stats["std_SAT"].max()]
largest_std_dev
```
